Columns to drop = Country( its all Kenyan data),Date of analysis,Total Country population,Level 1, Number.)

```python
df_ipc = df_ipc.drop(['Date of analysis','Country','Total country population','Level 1','Number'], axis=1)
df_ipc.head()
```

```python
# Standardizing columns
df_ipc['From'] = pd.to_datetime(df_ipc['From'])
df_ipc['Month'] = df_ipc['From'].dt.to_period('M')
df_ipc.head()
```

```python
df_ipc[df_ipc['Percentage'] > 0]
```

```python
# add target column 
df_ipc['in_Phase3+'] = (df_ipc['Percentage'] >= 0).astype(int)
df_ipc
```

```python
df_ipc = df_ipc.drop(['Validity period','From','To'],axis=1)
df_ipc
```

Exploring ipcphase data which contains older ipc data.

```python
ipc = pd.read_excel('./data/ipcphase.xlsx')
ipc.head()
```

```python
# taking only relevant columns
ipc = ipc[['geographic_unit_full_name','projection_start','projection_end','value']].copy()
ipc.head()
```

```python
# clean geographic column
ipc.info()
```

```python
ipc['Area'] = ipc['geographic_unit_full_name'].str.split(',').str[1].str.strip()
ipc.head(25)
```

```python
ipc['Area'].value_counts().tail(10)
```

```python
mask = ipc['Area'].isin(['High Potential Zone','Mixed Farming Zone','North Mugirango','Nyaribari Chache',   
'Nyaribari Masaba','Gilgil','Sigowet/Soin','Sirisia','Sotik','South Mugirango','Bahati','Endebess','Njoro',
'Nakuru Town East',"Nakuru Town West","Ol Kalou","Subukia",'Alego Usonga','Ugunja','Ugenya','Webuye East','Webuye West'])

ipc.loc[mask, 'Area'] = (
    ipc.loc[mask, 'geographic_unit_full_name']
    .str.rsplit(',', n=2)
    .str[-2]
    .str.strip()
)
```

```python
mask2 = ipc['Area'].isin([
    'Western','Kenya','Eastern','Central','Coast','Nyanza'])

ipc.loc[mask2,'Area'] = (ipc.loc[mask, 'geographic_unit_full_name']
        .str.split(',')
        .str[0]
        .str.strip())
```

```python
ipc = ipc[['Area'] + [c for c in ipc.columns if c != 'Area']]
ipc.head()
```

```python
ipc.info()
```

```python
ipc['projection_start'] = pd.to_datetime(ipc['projection_start'])
ipc['Month'] = ipc['projection_start'].dt.to_period('M')
ipc['in_Phase3+'] = (ipc['value'] >= 3).astype(int)
ipc.head()
```

```python
ipc.duplicated().sum()
```

```python
# inspecting duplicated rows
duplicated_mask = ipc.duplicated(subset=['Area','Month'], keep = False)
ipc[duplicated_mask]
```

```python
ipc = ipc.drop(['projection_start','projection_end'],axis=1)
```

```python
# Merging ipc and df_ipc
ipc_merged = ipc.merge(df_ipc,on=['Area','Month'],
                how='left',
                suffixes=['all','_r'])
```

```python
ipc_merged.info()
```

```python
ipc_merged = ipc_merged.drop(['in_Phase3+_r','Percentage','Phase'],axis=1)
ipc_merged.info()
```

```python
ipc_merged.isna().sum()
```

```python
ipc_merged[ipc_merged['Area'].isna()]
```

```python
ipc_merged = ipc_merged.dropna(subset=['Area'])
```

```python
ipc_merged.info()
```

```python
ipc_merged.rename(columns={'Area':'County','value':'IPC', 'in_Phase3+all':'in_Phase3+'},
                  inplace=True)
ipc_merged.head()
```
