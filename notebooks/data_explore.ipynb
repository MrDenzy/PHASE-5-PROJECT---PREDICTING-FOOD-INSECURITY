# Exploring and cleaning food prices data
I am going to focus only on maize(white) prices as it is a staple food in kenya and the best indicator for food security.

```python
df_fp = pd.read_csv('./data/wfp_food_prices_ken.csv')
df_fp.head(20)
```

```python
df_fp.info()
```

```python
df_fp = df_fp[df_fp['commodity_id'].isin(['51','67'])].copy()
df_fp.info()
```

```python
df_fp['commodity'].value_counts()
```

```python
df_fp
```

```python
df_p = df_fp[['date','admin2','market','unit','pricetype','price']].copy()
df_p.head()
```

```python
# checking to see if admin2 and market always have the same values
(df_p['admin2'] == df_p['market']).all()
```

```python
df_p[df_p['admin2'] != df_p['market']]
```

```python
df_p.info()
```

```python
df_p.isna().sum().sum()
```

```python
null_df = df_p[df_p['admin2'].isna()]
null_df['market'].value_counts()
```

```python
df_p['admin2'] = df_p['admin2'].fillna('Tana River')
df_p.info()
```

```python
# extracting numeric values from unit column to place in a separate column
df_p['unit_val'] = df_p['unit'].str.extract(r'(\d+)')
# fill null values from KG with 1
df_p['unit_val'] = df_p['unit_val'].fillna(1)

# turning unit_val and price columns to numeric
n_cols = ['unit_val','price']
df_p[n_cols]= df_p[n_cols].apply(pd.to_numeric,errors= 'coerce')
df_p.info()
```

```python
import numpy as np
df_p['price_per_kg']= (df_p['price']/df_p['unit_val']).replace(0,np.nan)
df_p['price_per_kg'].describe()
```

```python
df_p
```

```python
df_p = df_p[['date','admin2','price_per_kg']].copy()
df_p
```

```python
df_p.groupby(['admin2','date']).size()
```

```python
df_p.rename(columns={'admin2':'County',
                     'date':'Month',
                     'price_per_kg': 'maize_price_per_kg'}, inplace=True)
```

```python
df_p['Month'] = pd.to_datetime(df_p['Month'])
df_p['Month'] = df_p['Month'].dt.to_period('M')
month_counts = df_p.groupby('County')['Month'].nunique()
month_counts
```

```python
df_p = df_p.groupby(['County','Month'], as_index = False ).agg(
    {'maize_price_per_kg': 'mean'}
)
df_p
```

```python
df_p = df_p[df_p['Month'] >= pd.Period('2015-01',freq='M')].copy()
# creating a full county and monthly index
months = pd.period_range(df_p['Month'].min(),df_p['Month'].max(),freq = 'M')
counties = df_p['County'].unique()
full_index = pd.MultiIndex.from_product([counties,months], names=['County','Month'])

# reindexing
df_p = df_p.set_index(['County','Month']).reindex(full_index).reset_index()

df_p
```

```python

# forward filling missing prices 
df_p['maize_price_per_kg'] = df_p.groupby('County')['maize_price_per_kg'].ffill()
#back filling missing prices
df_p['maize_price_per_kg'] = df_p.groupby('County')['maize_price_per_kg'].bfill()

# adding lag features
df_p['price_lag1'] = df_p.groupby('County')['maize_price_per_kg'].shift(1)
df_p['price_lag2'] = df_p.groupby('County')['maize_price_per_kg'].shift(2)
df_p['price_pct_change'] = df_p.groupby('County')['maize_price_per_kg'].pct_change()
df_p
```

```python

df_p.head()
```

```python
df_p[48:]
```

```python
df_p['maize_price_per_kg'].isna().sum()
```

```python
df_p.info()
```

```python
df_p[['price_lag1','price_lag2','price_pct_change']] = df_p[['price_lag1','price_lag2','price_pct_change']].fillna(0) 
df_p.info()
```

# Exploring and Cleaning Conflict Data

```python
import pandas as pd
df_c = pd.read_excel('./data/Africa_aggregated_data_up_to-2026-02-07.xlsx')
df_c.head()
```

```python
# filtering to kenyan data only and relevant rows
df_c = df_c.loc[df_c['COUNTRY'].str.lower() == 'kenya',['WEEK','ADMIN1','EVENTS','FATALITIES']].copy()
df_c.head()
```

```python
df_c.info()
```

```python
df_c = (
    df_c
       .assign(Month = df_c['WEEK'].dt.to_period('M'))
       .groupby(['ADMIN1','Month'])
       .agg({'FATALITIES': 'sum','EVENTS': 'sum'})
       .reset_index())

df_c.info()
```

```python
df_c.head()
```

```python
# start with data from 2015 onwards
df_c = df_c.loc[df_c['Month'] >= pd.Period('2015-01',freq = 'M')].reset_index(drop=True).copy()
print(df_c.info())
print('\nConflict Sample:')
display(df_c.head(10))
```

```python
df_c.rename(columns={'ADMIN1':'County','FATALITIES':'Fatalities','EVENTS':'Events'},inplace=True)
df_c.head()
```

```python
df_c['County'].nunique()
```

```python
df_c.tail(30)
```

```python
# create a df with all county - month combinations covered.
months = pd.date_range(start='2015-01-01',end='2025-12',freq='MS').to_period('M')
counties = df_c['County'].unique()

# creating skeleton df
skl = pd.MultiIndex.from_product([months,counties],names=['Month','County']).to_frame(index=False)

# merging and filling
df_c = pd.merge(skl,df_c,on =['County','Month'] ,how='left')
# filling null values with 0
df_c[['Fatalities','Events']] = df_c[['Fatalities','Events']].fillna(0)
```
